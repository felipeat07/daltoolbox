```r
# DAL Library
# version 1.0.24

source("https://raw.githubusercontent.com/cefet-rj-dal/dal/main/examples/jupyter_daltoolbox.R")

#loading DAL
load_daltoolbox() 
loadlibrary("ggplot2")
loadlibrary("dplyr")
loadlibrary("reshape")
loadlibrary("RColorBrewer")
```

```r
#color palette
colors <- brewer.pal(4, 'Set1')

# setting the font size for all charts
font <- theme(text = element_text(size=16))
```

#### Synthetic time series

```r
x <- seq(0, 10, 0.25)
serie <- data.frame(x, sin=sin(x), cosine=cos(x)+5)
head(serie)
```

## Series plot

A series plot with a second y-axis is a particular case of a series plot with two different dependent variables. The first one is associated with the first y-axis. In contrast, the second one is associated with the second y-axis. 

```r
grf <- plot.series_second_axis(serie, colors=colors[1:2]) + font
plot(grf)
```
